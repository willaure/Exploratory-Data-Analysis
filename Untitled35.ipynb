```python
# Step 2: Load and Preview the Dataset¶
# Load the dataset from the provided URL. Use df.head() to display the first few rows to get an overview of the structure.
```

```python
# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Set pandas option to display all columns
pd.set_option('display.max_columns', None)

# Display the first few rows of the dataset
df.head()
```

```python
# Step 3: Handling Missing Data¶
# Identify and manage missing values in critical columns such as Employment, JobSat, and RemoteWork. Implement a strategy to fill or drop these values, depending on the significance of the missing data.
```

```python
# Step 3: Handling Missing Data

# First, let's check the missing values in our key columns
missing_data = df[['Employment', 'JobSat', 'RemoteWork']].isnull().sum()
print("Missing values before handling:")
print(missing_data)
print("\n")

# Strategy for each column:
# 1. Employment - Critical field, we'll drop rows with missing values
# 2. JobSat - Important but not critical, we'll fill with mode (most common value)
# 3. RemoteWork - Important, we'll fill with 'Not specified'

# Handle Employment - drop rows with missing values
df = df.dropna(subset=['Employment'])

# Handle JobSat - fill with mode (most frequent value)
job_sat_mode = df['JobSat'].mode()[0]
df['JobSat'] = df['JobSat'].fillna(job_sat_mode)

# Handle RemoteWork - fill with 'Not specified'
df['RemoteWork'] = df['RemoteWork'].fillna('Not specified')

# Verify the handling
missing_data_after = df[['Employment', 'JobSat', 'RemoteWork']].isnull().sum()
print("Missing values after handling:")
print(missing_data_after)

# Optional: Show summary of how many values were handled
print("\nHandling summary:")
print(f"- Dropped {missing_data['Employment'] - missing_data_after['Employment']} rows with missing Employment data")
print(f"- Filled {missing_data['JobSat'] - missing_data_after['JobSat']} JobSat values with mode: {job_sat_mode}")
print(f"- Filled {missing_data['RemoteWork'] - missing_data_after['RemoteWork']} RemoteWork values with 'Not specified'")
```

```python
# Step 4: Analysis of Experience and Job Satisfaction
# Analyze the relationship between years of professional coding experience (YearsCodePro) and job satisfaction (JobSat). Summarize YearsCodePro and calculate median satisfaction scores based on experience ranges.
# Create experience ranges for YearsCodePro (e.g., 0-5, 5-10, 10-20, >20 years).
# Calculate the median JobSat for each range.
# Visualize the relationship using a bar plot or similar visualization.
```

```python
# Step 4: Analysis of Experience and Job Satisfaction
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. First examine your JobSat values
print("Unique JobSat values in your data:")
print(df['JobSat'].unique())

# 2. Create a mapping based on what exists in your data
job_sat_values = df['JobSat'].dropna().unique()
job_sat_order = {
    'Very dissatisfied': 1,
    'Slightly dissatisfied': 2,
    'Neither satisfied nor dissatisfied': 3,
    'Slightly satisfied': 4,
    'Very satisfied': 5
}

# Only keep mappings that exist in your data
job_sat_order = {k: v for k, v in job_sat_order.items() if k in job_sat_values}

# If no standard values found, create numeric mapping from scratch
if not job_sat_order:
    unique_values = sorted(df['JobSat'].dropna().unique())
    job_sat_order = {v: i+1 for i, v in enumerate(unique_values)}

print("\nUsing this mapping:")
print(job_sat_order)

# Create a copy of the dataframe to avoid warnings
df_analysis = df.copy()

# Handle missing values
df_analysis = df_analysis.dropna(subset=['YearsCodePro', 'JobSat'])

# Convert YearsCodePro to numeric
df_analysis['YearsCodePro'] = df_analysis['YearsCodePro'].replace({
    'Less than 1 year': '0.5',
    'More than 50 years': '50'
}).astype(float)

# Create experience bins
bins = [0, 5, 10, 20, float('inf')]
labels = ['0-5 years', '5-10 years', '10-20 years', '20+ years']
df_analysis['ExperienceRange'] = pd.cut(
    df_analysis['YearsCodePro'],
    bins=bins,
    labels=labels,
    right=False
)

# Map JobSat to numeric values
df_analysis['JobSatNumeric'] = df_analysis['JobSat'].map(job_sat_order)

# Calculate median satisfaction
median_satisfaction = df_analysis.groupby('ExperienceRange', observed=True)['JobSatNumeric'].median().reset_index()
median_satisfaction['MedianSatisfaction'] = median_satisfaction['JobSatNumeric'].map(
    {v: k for k, v in job_sat_order.items()}
)

print("\nMedian Job Satisfaction by Experience Range:")
print(median_satisfaction[['ExperienceRange', 'MedianSatisfaction', 'JobSatNumeric']])

# Visualization
plt.figure(figsize=(10, 6))
ax = sns.barplot(
    x='ExperienceRange',
    y='JobSatNumeric',
    data=df_analysis,
    estimator=np.median,
    ci=None,
    order=labels
)
plt.title('Median Job Satisfaction by Professional Coding Experience', fontsize=14)
plt.xlabel('Years of Professional Coding Experience', fontsize=12)
plt.ylabel('Median Satisfaction Score', fontsize=12)

# Add labels to bars
for index, row in median_satisfaction.iterrows():
    if not pd.isna(row['JobSatNumeric']):
        ax.text(
            index,
            row['JobSatNumeric'] + 0.1,
            f"{row['MedianSatisfaction']}\n({row['JobSatNumeric']:.1f})",
            ha='center',
            color='black',
            fontsize=10
        )

plt.ylim(0, max(job_sat_order.values()) + 0.5)
plt.tight_layout()
plt.show()
```

```python
# Step 6: Analyzing Remote Work Preferences by Job Role
# Analyze trends in remote work based on job roles. Use the RemoteWork and Employment columns to explore preferences and examine if specific job roles prefer remote work more than others.
# Use a count plot to show remote work distribution.
# Cross-tabulate remote work preferences by employment type (e.g., full-time, part-time) and job roles.
```

```python
# Step 6: Analyzing Remote Work Preferences by Job Role - Fixed Version

# 1. Data preparation
df_remote = df.dropna(subset=['RemoteWork', 'Employment']).copy()

# Standardize remote work categories (adjust based on your actual data)
remote_mapping = {
    'Fully remote': 'Remote',
    'Hybrid (some remote, some in-person)': 'Hybrid',
    'Full in-person': 'In-office'
}
# Only apply mapping for values that exist in the data
existing_categories = {k: v for k, v in remote_mapping.items() if k in df_remote['RemoteWork'].unique()}
df_remote['RemoteWork'] = df_remote['RemoteWork'].replace(existing_categories)

# 2. Cross-tabulation analysis (more robust version)
print("Remote Work Preference by Employment Type:")

# Create cross-tab without margins first to check
cross_tab = pd.crosstab(
    index=df_remote['Employment'],
    columns=df_remote['RemoteWork'],
    normalize='index'
).round(3) * 100

# Sort by total respondents and show top 10
top_roles = df_remote['Employment'].value_counts().head(10).index
cross_tab_top = cross_tab.loc[top_roles]

# Display with styling
display(
    cross_tab_top
    .style
    .background_gradient(cmap='Blues', axis=1)
    .format('{:.1f}%')
)

# 3. Visualization
plt.figure(figsize=(12, 8))

# Get the top 5 roles for clearer visualization
top_5_roles = df_remote['Employment'].value_counts().head(5).index
df_top_5 = df_remote[df_remote['Employment'].isin(top_5_roles)]

sns.countplot(
    y='Employment',
    hue='RemoteWork',
    data=df_top_5,
    order=top_5_roles,
    hue_order=['Remote', 'Hybrid', 'In-office'],
    palette='coolwarm'
)
plt.title('Remote Work Preferences by Top 5 Job Roles', fontsize=14)
plt.xlabel('Number of Respondents')
plt.ylabel('Job Role')
plt.legend(title='Work Preference', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 4. Additional analysis - Remote work preference percentages
print("\nPercentage Remote Work by Job Role:")
remote_percentage = df_remote.groupby('Employment')['RemoteWork']\
    .apply(lambda x: (x == 'Remote').mean() * 100)\
    .sort_values(ascending=False)\
    .head(10)
print(remote_percentage.round(1))
```

```python
# Step 7: Analyzing Programming Language Trends by Region¶

# Analyze the popularity of programming languages by region. Use the LanguageHaveWorkedWith column to investigate which languages are most used in different regions.

# Filter data by country or region.

# Visualize the top programming languages by region with a bar plot or heatmap.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data (in case it's not already loaded)
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Keep only relevant columns
df_lang = df[['Country', 'LanguageHaveWorkedWith']].dropna()

# Split multiple languages into separate rows
df_exploded = df_lang.assign(Language=df_lang['LanguageHaveWorkedWith'].str.split(';')).explode('Language')

# Group by Country and Language to count occurrences
lang_by_country = df_exploded.groupby(['Country', 'Language']).size().reset_index(name='Count')

# Pivot the data to get a matrix: rows = countries, columns = languages
lang_matrix = lang_by_country.pivot(index='Country', columns='Language', values='Count').fillna(0)

# OPTIONAL: Normalize by row to get percentages (uncomment if needed)
# lang_matrix = lang_matrix.div(lang_matrix.sum(axis=1), axis=0)

# Limit to top 10 countries (by number of respondents)
top_countries = df['Country'].value_counts().head(10).index
lang_matrix_top = lang_matrix.loc[top_countries]

# Limit to top 10 languages
top_languages = df_exploded['Language'].value_counts().head(10).index
lang_matrix_top = lang_matrix_top[top_languages]

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(lang_matrix_top, annot=True, fmt=".0f", cmap="YlGnBu", cbar_kws={'label': 'Respondent Count'})
plt.title('Top Programming Languages by Country (Stack Overflow Survey)')
plt.ylabel('Country')
plt.xlabel('Programming Language')
plt.tight_layout()
plt.show()
```

```python
# Step 8: Correlation Between Experience and Satisfaction¶
# Examine how years of experience (YearsCodePro) correlate with job satisfaction (JobSatPoints_1). Use a scatter plot to visualize this relationship.
```

```python
# Step 8: Correlation Between Experience and Satisfaction
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 1. Data Preparation
df_corr = df.copy()

# Clean YearsCodePro - handle categorical responses and convert to numeric
df_corr['YearsCodePro'] = df_corr['YearsCodePro'].replace({
    'Less than 1 year': 0.5,
    'More than 50 years': 50
}).astype(float)

# Drop rows with missing values in our key columns
df_corr = df_corr.dropna(subset=['YearsCodePro', 'JobSatPoints_1'])

# 2. Scatter Plot with Regression Line
plt.figure(figsize=(12, 8))

# Create the scatter plot with regression line
sns.regplot(
    x='YearsCodePro',
    y='JobSatPoints_1',
    data=df_corr,
    scatter_kws={'alpha': 0.4, 'color': 'steelblue'},
    line_kws={'color': 'red', 'linewidth': 2},
    lowess=False  # Set to True for local regression if pattern is non-linear
)

# 3. Calculate Correlation Coefficients
pearson_corr, pearson_p = stats.pearsonr(df_corr['YearsCodePro'], df_corr['JobSatPoints_1'])
spearman_corr, spearman_p = stats.spearmanr(df_corr['YearsCodePro'], df_corr['JobSatPoints_1'])

# 4. Add Annotations and Formatting
plt.title('Relationship Between Professional Experience and Job Satisfaction', fontsize=16, pad=20)
plt.xlabel('Years of Professional Coding Experience', fontsize=12)
plt.ylabel('Job Satisfaction Score', fontsize=12)
plt.grid(alpha=0.2)

# Add correlation info to plot
textstr = '\n'.join((
    f'Pearson r = {pearson_corr:.2f} (p = {pearson_p:.3f})',
    f'Spearman ρ = {spearman_corr:.2f} (p = {spearman_p:.3f})'
))
plt.gca().text(
    0.05, 0.95, textstr,
    transform=plt.gca().transAxes,
    fontsize=12,
    verticalalignment='top',
    bbox=dict(boxstyle='round', facecolor='white', alpha=0.8)
)

plt.tight_layout()
plt.show()

# 5. Additional Analysis - Binned Experience Analysis
print("\nAdditional Analysis - Satisfaction by Experience Bins:")

# Create experience bins
bins = [0, 2, 5, 10, 20, np.inf]
labels = ['0-2', '2-5', '5-10', '10-20', '20+']
df_corr['ExpBin'] = pd.cut(df_corr['YearsCodePro'], bins=bins, labels=labels)

# Calculate median satisfaction by bin
bin_stats = df_corr.groupby('ExpBin')['JobSatPoints_1'].agg(['median', 'count'])
print(bin_stats)

# 6. Boxplot Visualization
plt.figure(figsize=(10, 6))
sns.boxplot(
    x='ExpBin',
    y='JobSatPoints_1',
    data=df_corr,
    order=labels,
    palette='viridis'
)
plt.title('Job Satisfaction Distribution by Experience Level', fontsize=14)
plt.xlabel('Years of Professional Experience', fontsize=12)
plt.ylabel('Job Satisfaction Score', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Step 9: Educational Background and Employment Type
# Explore how educational background (EdLevel) relates to employment type (Employment). Use cross-tabulation and visualizations to understand if higher education correlates with specific employment types.
```

```python
# Step 9: Educational Background and Employment Type Analysis - Fixed Version

# 1. Data Preparation
df_edu = df.copy()

# Clean and standardize education levels (adjust based on your actual data)
education_mapping = {
    'No formal education': 'No formal education',
    'Primary/elementary school': 'Primary school',
    'Secondary school': 'Secondary school',
    'Some college/university': 'Some college',
    'Associate degree': 'Associate degree',
    'Bachelor’s degree': "Bachelor's degree",
    'Master’s degree': "Master's degree",
    'Professional degree': 'Professional degree',
    'Doctoral degree': 'Doctoral degree'
}

# Apply mapping only to existing categories
existing_edu = {k: v for k, v in education_mapping.items() if k in df_edu['EdLevel'].unique()}
df_edu['EdLevel'] = df_edu['EdLevel'].replace(existing_edu)

# 2. Cross-tabulation with percentages (fixed version)
print("Education Level by Employment Type (% within each employment type):")

# Create cross-tabulation with only non-null values
cross_tab = pd.crosstab(
    index=df_edu['Employment'].dropna(),
    columns=df_edu['EdLevel'].dropna(),
    normalize='index'
).round(3) * 100

# Display with styling (show top 10 employment types)
top_employment = df_edu['Employment'].value_counts().head(10).index
display(
    cross_tab
    .loc[top_employment.intersection(cross_tab.index)]  # Only show existing indices
    .style
    .background_gradient(cmap='Blues', axis=1)
    .format('{:.1f}%')
    .set_caption("Percentage Distribution of Education Levels Within Each Employment Type")
)

# 3. Stacked Bar Chart Visualization (fixed version)
plt.figure(figsize=(14, 8))

# Get top 5 employment types that exist in our data
top_employment = df_edu['Employment'].value_counts().head(5).index
df_top = df_edu[df_edu['Employment'].isin(top_employment)]

# Create ordered education levels (only those present in data)
edu_order = [
    'No formal education',
    'Primary school',
    'Secondary school',
    'Some college',
    'Associate degree',
    "Bachelor's degree",
    "Master's degree",
    'Professional degree',
    'Doctoral degree'
]
edu_order = [x for x in edu_order if x in df_top['EdLevel'].unique()]

# Create plot only if we have data to plot
if len(edu_order) > 0 and len(top_employment) > 0:
    ax = pd.crosstab(
        index=df_top['Employment'],
        columns=df_top['EdLevel']
    ).loc[top_employment, edu_order].plot(
        kind='barh',
        stacked=True,
        figsize=(14, 8),
        cmap='viridis'
    )

    plt.title('Educational Background by Employment Type', fontsize=16, pad=20)
    plt.xlabel('Number of Respondents', fontsize=12)
    plt.ylabel('Employment Type', fontsize=12)
    plt.legend(title='Education Level', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(axis='x', alpha=0.3)

    # Add percentage annotations if we have data
    if len(df_top) > 0:
        total_counts = df_top['Employment'].value_counts()
        for bars in ax.containers:
            labels = []
            for bar in bars:
                if bar.get_width() > 0:
                    emp_type = bar.get_y() + bar.get_height()/2
                    emp_type = ax.get_yticklabels()[int(emp_type)].get_text()
                    percentage = 100 * bar.get_width() / total_counts[emp_type]
                    labels.append(f'{percentage:.1f}%')
                else:
                    labels.append('')
            ax.bar_label(bars, labels=labels, label_type='center', fontsize=9)

    plt.tight_layout()
    plt.show()
else:
    print("Not enough data to create visualization. Check your EdLevel and Employment columns.")

# 4. Statistical Summary - Fixed version
print("\nMost Common Education Level for Each Employment Type:")

# Safe mode calculation that handles empty groups
def safe_mode(series):
    try:
        return series.mode()[0]
    except:
        return pd.NA

# Get most common education for each employment type
common_edu = df_edu.groupby('Employment')['EdLevel'].agg(safe_mode).dropna()

if len(common_edu) > 0:
    display(
        common_edu
        .value_counts()
        .to_frame('Number of Employment Types')
        .rename_axis('Most Common Education Level')
        .style.background_gradient(cmap='Greens')
    )
else:
    print("No valid data to calculate most common education levels.")
```
